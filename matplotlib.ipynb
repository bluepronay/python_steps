```python
import matplotlib.pyplot as plt

# Create a new figure
fig = plt.figure()

# Add a set of axes to the figure
ax = fig.add_axes([0, 0, 1, 1])

# Define the data
languages = ['Hindi', 'English', 'Sanskrit', 'Tamil', 'Bengali']
persons = [84, 99, 150, 33, 76]  

# Create a bar chart
ax.bar(languages, persons)

# Add labels to the x and y axes
plt.xlabel('Languages')
plt.ylabel('No of Persons')

# Display the plot
plt.show()
```

Pie Chart

```python
import matplotlib.pyplot as plt

# Create a new figure
fig1 = plt.figure()

# Add a set of axes to the figure covering the entire area
ax = fig1.add_axes([0, 0, 1, 1])

# Create a pie chart on the axes with labels and percentages
ax.pie(persons, labels=languages, autopct='%1.2f%%')
#autopct means one value of float after no

# Display the plot
plt.show()
```

Scatter Plot

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)

# Creating a new figure for plotting
fig2 = plt.figure()

# Adding a set of axes to the figure. The list [0, 0, 1, 1] specifies the position and size of the axes
# [left, bottom, width, height] in fractions of figure width and height
ax = fig2.add_axes([0, 0, 1, 1])

# Creating a scatter plot of x vs y with green color
ax.scatter(x, y, color='g')

# Creating a scatter plot of x vs z with red color
ax.scatter(x, z, color='r')

# Displaying the plot
plt.show()
```

3D Scatter plot

```python
# Creating a new figure for plotting
fig3 = plt.figure()

# Adding a 3D set of axes to the figure
ax = fig3.add_subplot(111, projection='3d')

# Generating an array of 100 random values between 0 and 20
z = 20 * np.random.random(100)

# Computing the sine of each value in the array z
x = np.sin(z)

# Computing the cosine of each value in the array z
y = np.cos(z)

# Creating a 3D scatter plot of x, y, and z with color mapping based on z values and using the 'Blues' colormap
ax.scatter(x, y, z, c=z, cmap='Blues')

# Displaying the plot
plt.show()
```
